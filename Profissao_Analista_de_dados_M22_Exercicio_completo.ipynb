# **Exercícios**

## 1\. Pinguins

Neste exercício, vamos utilizar uma base de dados com informações sobre penguins. A idéia é preparar a base de dados para prever a espécie do penguin (variável resposta) baseado em suas características físicas e geográficas (variáveis preditivas).

```
import numpy as np
import pandas as pd
import seaborn as sns
```

```
data = sns.load_dataset('penguins')
df = pd.DataFrame(data)
```

```
print(df)
```

```
df.head()
```

### **1.1. Valores nulos**

A base de dados possui valores faltantes, utilize os conceitos da aula para trata-los.

```
# resposta da questão 1.1
df.isnull().sum()
```

```
display(df[df.isnull().any(axis=1)])
```

```
df['bill_length_mm'] = df['bill_length_mm'].fillna(df['bill_length_mm'].mean())
df['bill_depth_mm'] = df['bill_depth_mm'].fillna(df['bill_depth_mm'].mean())
df['flipper_length_mm'] = df['flipper_length_mm'].fillna(df['flipper_length_mm'].mean())
df['body_mass_g'] = df['body_mass_g'].fillna(df['body_mass_g'].mean())
df = df.drop('sex', axis=1)
```

```
df.isnull().sum()
```

```
display(df[df.isnull().any(axis=1)])
```

### **1.2. Variáveis numéricas**

Identifique as variáveis numéricas e crie uma nova coluna **padronizando** seus valores. A nova coluna deve ter o mesmo nome da coluna original acrescidade de "*_std*".

> **Nota**: Você não deve tratar a variável resposta.

```
# resposta da questão 1.2
df['bill_length_mm_std'] = df['bill_length_mm'].apply(lambda valor: (valor - df['bill_length_mm'].mean()) / df['bill_length_mm'].std() )
df['bill_depth_mm_std'] = df['bill_depth_mm'].apply(lambda valor: (valor - df['bill_depth_mm'].mean()) / df['bill_depth_mm'].std() )
df['flipper_length_mm_std'] = df['flipper_length_mm'].apply(lambda valor: (valor - df['flipper_length_mm'].mean()) / df['flipper_length_mm'].std() )
df['body_mass_g_std'] = df['body_mass_g'].apply(lambda valor: (valor - df['body_mass_g'].mean()) / df['body_mass_g'].std() )

df.head()
```

### **1.3. Variáveis categóricas**

Identifique as variáveis categóricas nominais e ordinais, crie uma nova coluna aplicando a técnica correta de conversão a seus valores. A nova coluna deve ter o mesmo nome da coluna original acrescidade de "*_nom*" ou "*_ord*".

> **Nota**: Você não deve tratar a variável resposta.

```
# resposta da questão 1.3
df['island'].drop_duplicates()
```

```

df['island_nom_Torgersen'] = df['island'].apply(lambda valor: 1 if valor == 'Torgersen' else 0)
df['island_nom_Biscoe'] = df['island'].apply(lambda valor: 1 if valor == 'Biscoe' else 0)
df['island_nom_Dream'] = df['island'].apply(lambda valor: 1 if valor == 'Dream' else 0)

df.head()
```

### **1.4. Limpeza**

Descarte as colunas originais e mantenha apenas a variável resposta e as variáveis preditivas com o sufixo *_std*", *_nom*" e "*_ord*".

```
# resposta da questão 1.4
df.columns
```

```
df = df[['species', 'bill_length_mm_std',
             'bill_depth_mm_std', 'flipper_length_mm_std',
             'body_mass_g_std','island_nom_Torgersen',
             'island_nom_Biscoe', 'island_nom_Dream'
]]


df.head()
```
